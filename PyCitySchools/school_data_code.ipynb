```python
import pandas as pd
import numpy as np
```

```python
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
school_list=school_data_complete["school_name"].unique()
total_schools= len(school_data_complete["school_name"].unique())
total_students= len(school_data_complete['student_name'])
total_budget= sum(school_data_complete["budget"].unique())
avg_math= school_data_complete['math_score'].mean()
avg_reading= school_data_complete['reading_score'].mean()
math_pass= len(school_data_complete.loc[school_data_complete["math_score"]>= 70,:])/total_students*100
reading_pass= len(school_data_complete.loc[school_data_complete["reading_score"]>= 70,:])/total_students*100
overall_pass= len(school_data_complete.loc[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70), :])/total_students*100
district_summary= [{"Total Schools": total_schools, "Total Students": total_students, "Total Budget": total_budget, "Average Math Score": avg_math, \
                  'Average Reading Score':avg_reading, "% Passing Math": math_pass, "% Passing Reading": reading_pass, "% Overall Passing": overall_pass}]
district_summary_pd= pd.DataFrame(district_summary)
district_summary_pd
```

```python
school_avg_df= school_data_complete[["school_name","reading_score","math_score"]]
school_avg_df.loc[(school_avg_df["math_score"]>=70) & (school_avg_df["reading_score"]>=70), 'overall_pass']= 1
school_avg_df.loc[(school_avg_df["math_score"]<70) | (school_avg_df["reading_score"]<70), 'overall_pass']= 0
school_avg_df.loc[(school_avg_df["math_score"]>=70), 'math_pass']= 1
school_avg_df.loc[(school_avg_df["math_score"]<70), 'math_pass']= 0
school_avg_df.loc[(school_avg_df["reading_score"]>=70), 'reading_pass']= 1
school_avg_df.loc[(school_avg_df["reading_score"]<70), 'reading_pass']= 0
grouped= school_avg_df.groupby("school_name", as_index= False).agg('mean')
grouped['overall_pass']= grouped['overall_pass']*100
grouped['math_pass']= grouped['math_pass']*100
grouped['reading_pass']= grouped['reading_pass']*100
school_data_grouped = pd.merge(school_data, grouped, how="left", on=["school_name", "school_name"])
school_data_grouped
del school_data_grouped['School ID']
school_level=school_data_grouped.set_index('school_name')
school_level['per_student']= school_level['budget']/school_level['size']
school_lvl =school_level[['type','size','budget','per_student','reading_score','math_score','overall_pass']]
school_level2 = school_lvl.rename(columns={"type": "School Type",
                                               "reading_score": "Average Reading Score",
                                               "math_score": "Average Math Score",
                                               "size": "Total Students",
                                               "budget": "Total School Budget",
                                               "per_student": "Total School Budget",
                                               "overall_pass": "% Overall Passing"})

school_level2.index.names=['']
school_level2
```

```python
top_five=school_level2.sort_values(by=['% Overall Passing'], ascending=False)
top_five.head()
```

```python
bottom_five= school_level2.sort_values(by=['% Overall Passing'])
bottom_five.head()
```

```python
grade_avg_df= school_data_complete[["school_name",'grade',"reading_score","math_score"]]
```

```python
grade_avg_df.loc[(grade_avg_df["grade"]=='9th'),'9th']= grade_avg_df["math_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='10th'),'10th']= grade_avg_df["math_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='11th'),'11th']= grade_avg_df["math_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='12th'),'12th']= grade_avg_df["math_score"]
grades_grouped= grade_avg_df.groupby("school_name").agg('mean')
math=grades_grouped
del math['math_score']
math
del math['reading_score']
math.index.names=['']
math
```

```python
grade_avg_df.loc[(grade_avg_df["grade"]=='9th'),'9th']= grade_avg_df["reading_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='10th'),'10th']= grade_avg_df["reading_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='11th'),'11th']= grade_avg_df["reading_score"]
grade_avg_df.loc[(grade_avg_df["grade"]=='12th'),'12th']= grade_avg_df["reading_score"]
grades_grouped= grade_avg_df.groupby("school_name").agg('mean')
reading=grades_grouped
del reading['math_score']
reading
del reading['reading_score']
reading.index.names=['']
reading
```

```python
per_student_bins=[0,584,629,644,680]
labels= ("<$585", '$585-630', '$630-645', '$645-680')
school_level["per_student_bins"]=pd.cut(school_level['per_student'], per_student_bins, labels= labels)
per_student_df=school_level[['per_student_bins','reading_score','math_score','math_pass','reading_pass','overall_pass']]
ps_df = per_student_df.rename(columns={"per_student_bins": "Spending Ranges (Per Student)",
                                               "reading_score": "Average Reading Score",
                                               "math_score": "Average Math Score",
                                               "math_pass": "% Passing Math",
                                               "reading_pass": "% Passing Reading",
                                               "overall_pass": "% Overall Passing"})
cost_bins=ps_df.groupby('Spending Ranges (Per Student)').agg('mean')
cost_bins
```

```python
size_bins=[0,999,1999,5000]
size_labels= ("Small (<1000)", 'Medium (1000-2000)', 'Large (2000-5000)')
school_level["size_bins"]=pd.cut(school_level['size'], size_bins, labels= size_labels)
size_df=school_level[['size_bins','reading_score','math_score','math_pass','reading_pass','overall_pass']]
sized_df = size_df.rename(columns={"size_bins": "School Size",
                                               "reading_score": "Average Reading Score",
                                               "math_score": "Average Math Score",
                                               "math_pass": "% Passing Math",
                                               "reading_pass": "% Passing Reading",
                                               "overall_pass": "% Overall Passing"})
sized_df.groupby("School Size").agg('mean')
```
